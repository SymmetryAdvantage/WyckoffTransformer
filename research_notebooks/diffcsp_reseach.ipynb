```python
test_structure = diffcsp.data[diffcsp.data.structure.map(len) == 20].structure.iloc[2]
```

```python
from pymatgen.core.lattice import Lattice
from pymatgen.core.structure import Structure

# Use the average lattice parameter and angle from test_structure for a perfect rhombohedral lattice
a_rhomb = (test_structure.lattice.a + test_structure.lattice.b + test_structure.lattice.c) / 3
alpha_rhomb = (test_structure.lattice.alpha + test_structure.lattice.beta + test_structure.lattice.gamma) / 3

# Create a new rhombohedral lattice
rhombohedral_lattice = Lattice.from_parameters(
    a_rhomb, a_rhomb, a_rhomb,
    alpha_rhomb, alpha_rhomb, alpha_rhomb
)

# Build a new structure with the same species and fractional coordinates
ideal_test_structure = Structure(
    rhombohedral_lattice,
    test_structure.species,
    test_structure.frac_coords,
    coords_are_cartesian=False
)
```

```python
from pymatgen.symmetry.groups import SpaceGroup
from pymatgen.core.structure import Structure
import numpy as np
from itertools import combinations

# 1. Iterate over all space groups which have rhombohedral lattice
# Rhombohedral space groups are trigonal and their symbol starts with 'R'
rhombohedral_sgs = []
for i in range(1, 231):
    sg = SpaceGroup.from_int_number(i)
    if sg.crystal_system == 'trigonal' and sg.symbol.startswith('R'):
        rhombohedral_sgs.append(i)

symmetric_structures = []
# 2. For each space group apply all the transformations to the structure
for sg_num in rhombohedral_sgs:
    spacegroup = SpaceGroup.from_int_number(sg_num)
    symm_ops = spacegroup.symmetry_ops

    new_sites = []
    for op in symm_ops:
        for site in test_structure:
            new_site = op.operate(site.frac_coords)
            new_sites.append(new_site)

    # Create a new structure with all the transformed sites
    temp_structure = Structure(
        test_structure.lattice,
        [site.specie for site in test_structure] * len(symm_ops),
        new_sites,
        coords_are_cartesian=False
    )

    # 3. Merge atoms which are closer than a threshold in fractional coords
    # We do this manually since merge_sites uses Cartesian distances.
    tol = 0.1
    to_remove = []
    
    # Get a list of sites to work with
    sites = list(temp_structure)
    
    for i, j in combinations(range(len(sites)), 2):
        if i not in to_remove and j not in to_remove:
            # Get distance in fractional coordinates, accounting for periodicity
            frac_dist = np.abs(sites[i].frac_coords - sites[j].frac_coords)
            frac_dist -= np.round(frac_dist)
            
            if np.all(np.abs(frac_dist) < tol) and sites[i].specie == sites[j].specie:
                to_remove.append(j)

    if to_remove:
        temp_structure.remove_sites(sorted(list(set(to_remove))))

    # 4. Save such structures into a list
    symmetric_structures.append(temp_structure)

print(f"Generated {len(symmetric_structures)} structures for {len(rhombohedral_sgs)} rhombohedral space groups.")
# You can inspect the list `symmetric_structures` to see the results.
# For example, to see the first generated structure:
# symmetric_structures[0]
```

```python
# Find 5 structures from the training set that belong to rhombohedral space groups
rhombohedral_structures_df = mp_20_train.data[mp_20_train.data['group'].isin(rhombohedral_sgs)]
test_structures_rhombohedral = rhombohedral_structures_df.structure.head(5).tolist()
original_sgs = rhombohedral_structures_df.group.head(5).tolist()

all_symmetric_structures = []

for test_structure in test_structures_rhombohedral[:5]:
    symmetric_structures = []
    # For each space group apply all the transformations to the structure
    for sg_num in rhombohedral_sgs:
        spacegroup = SpaceGroup.from_int_number(sg_num)
        symm_ops = spacegroup.symmetry_ops

        new_sites = []
        for op in symm_ops:
            for site in test_structure:
                new_site = op.operate(site.frac_coords)
                new_sites.append(new_site)

        # Create a new structure with all the transformed sites
        temp_structure = Structure(
            test_structure.lattice,
            [site.specie for site in test_structure] * len(symm_ops),
            new_sites,
            coords_are_cartesian=False
        )

        # Merge atoms which are closer than a threshold in fractional coords
        tol = 0.1
        to_remove = []
        
        sites = list(temp_structure)
        
        for i, j in combinations(range(len(sites)), 2):
            if i not in to_remove and j not in to_remove:
                frac_dist = np.abs(sites[i].frac_coords - sites[j].frac_coords)
                frac_dist -= np.round(frac_dist)
                
                if np.all(np.abs(frac_dist) < tol) and sites[i].specie == sites[j].specie:
                    to_remove.append(j)

        if to_remove:
            temp_structure.remove_sites(sorted(list(set(to_remove))))

        symmetric_structures.append(temp_structure)
    all_symmetric_structures.append(symmetric_structures)

# Print the number of atoms for each generated structure for each test case
for i, (original_sg, structures) in enumerate(zip(original_sgs, all_symmetric_structures)):
    print(f"\nTest Structure {i+1} (Original SG: {original_sg}):")
    for j, s in enumerate(structures):
        print(f"  - With SG {rhombohedral_sgs[j]}: {len(s)} atoms")
```

```python
from pyxtal import pyxtal
from pyxtal.msg import Comp_CompatibilityError

print("Generating ideal structures with pyXtal and analyzing them...")

# Store results in a dictionary for clarity
analysis_results = {}

for _, record in rhombohedral_structures_df.iloc[:5].iterrows():
    s = pyxtal()
    s.from_random(dim=3, group=record.group, species=record.species, numIons=record.numIons)
    
    test_structure = s.to_pymatgen()
    print(f"\n--- Analyzing structure for SG {sg_num} ({len(test_structure)} atoms) ---")

    symmetric_structures = []
    for analysis_sg_num in rhombohedral_sgs:
        spacegroup = SpaceGroup.from_int_number(analysis_sg_num)
        symm_ops = spacegroup.symmetry_ops

        new_sites = []
        for op in symm_ops:
            for site in test_structure:
                new_site = op.operate(site.frac_coords)
                new_sites.append(new_site)

        temp_structure = Structure(
            test_structure.lattice,
            [site.specie for site in test_structure] * len(symm_ops),
            new_sites,
            coords_are_cartesian=False
        )

        tol = 0.01
        to_remove = []
        sites = list(temp_structure)
        
        for i, j in combinations(range(len(sites)), 2):
            if i not in to_remove and j not in to_remove:
                frac_dist = np.abs(sites[i].frac_coords - sites[j].frac_coords)
                frac_dist -= np.round(frac_dist)
                
                if np.all(np.abs(frac_dist) < tol) and sites[i].specie == sites[j].specie:
                    to_remove.append(j)

        if to_remove:
            temp_structure.remove_sites(sorted(list(set(to_remove))))
        
        symmetric_structures.append(temp_structure)
    
    analysis_results[sg_num] = symmetric_structures
    
    # Print results for the current structure
    for j, struct in enumerate(symmetric_structures):
        match_str = "<- Should Match" if rhombohedral_sgs[j] == sg_num else ""
        print(f"  - With SG {rhombohedral_sgs[j]}: {len(struct)} atoms {match_str}")

print("\n\nAnalysis complete.")
```
