```python
import gzip
import pickle
with gzip.open("cache/matbench_discovery_mp_2022/data.pkl.gz", "rb") as f:
    dataset_pd = pickle.load(f)
```

```python
import pandas as pd
mp_2022 = pd.concat([dataset_pd["train"], dataset_pd["val"]], axis=0)
```

```python
wp_counts_mp_2022 = mp_2022.site_symmetries.map(len)
```

```python
wp_counts_wbm = dataset_pd['test'].site_symmetries.map(len)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
bins=np.linspace(0, 360, 100)
ax.hist(wp_counts_mp_2022, bins=bins, alpha=0.5, label="MP 2022", density=True)
ax.hist(wp_counts_wbm, bins=bins, alpha=0.5, label="WBM", density=True)
ax.set_xlabel("Number of Wyckoff positions")
ax.set_ylabel("Density")
ax.legend();
```
